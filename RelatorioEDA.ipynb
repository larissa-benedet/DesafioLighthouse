Descrevendo graficamente os dados do segundo *dataset* chamado de *wage_test*, com gráficos boxplot e gráficos de barras, para assim poder visualizar tanto a distribuição massiva dos dados quanto seus outliers. 

Junto aos gráfico estão suas estatísticas descritivas correspondentes, sendo elas a média aritmética, pois é uma medida de posição que indica um valor uniforme dos dados; a moda, pois retorna o valor com maior frequência nos dados; e o desvio padrão, pois indica uma medida de dispersão dos dados em torno de média, isto é, um grande desvio padrão indica que os dados estão espalhados longe da média e um pequeno desvio padrão indica que os dados estão agrupados perto da média.

```python
#Importando bibliotecas.
import numpy as np
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
#Lendo o dataset de teste.
wage_test = pd.read_csv('wage_test.csv', 
sep=r'\s*,\s*', 
engine='python', 
na_values='?')
wage_test.head()
```

```python
#Variável 'age'.
sb.boxplot(x='age', data=wage_test, color='purple')
plt.title('Distribuição da variável "age"')
plt.show()

print('A média aritmética da variável "age" é: ', int(wage_test['age'].mean()))
print('A moda da variável "age" é: ', int(wage_test['age'].mode()))
print('O desvio padrão da variável "age" é: ', float(wage_test['age'].std()))
```

```python
#Variável 'workclass'.
wage_test['workclass'].value_counts().plot(kind='bar', figsize=(20,8), width=0.8, color=['#FF8C00','#32CD32','#BC8F8F','#9932CC','#8B0000','#87CEEB','#FFD700','#FF69B4'])
plt.title('Distribuição da variável "workclass"')
plt.show()

print('A média aritmética da variável "workclass" é: ', wage_test['workclass'].value_counts().mean())
print('A moda da variável "workclass" é: ', wage_test['workclass'].mode())
print('O desvio padrão da variável "workclass" é: ', wage_test['workclass'].value_counts().std())
```
